```python
"""
----------

Jason

----------

"""

class AbstractSimulation:
    def __init__(self, number_steps):
        self.number_steps = number_steps
        self.show_intermediate_steps = True

    def initialize_sim(self):
        pass

    def run_one_step(self):
        raise NotImplementedError

    def print_sim_state(self):
        pass

    def run(self):
        self.initialize_sim()
        for a in range(self.number_steps):
            self.run_one_step()
            if self.show_intermediate_steps:
                self.print_sim_state()
        if not (self.show_intermediate_steps):
            self.print_sim_state()


class Rule90(AbstractSimulation):
    def __init__(self, number_steps, array):
        super().__init__(number_steps)
        self.array = array
        self.steps = 0

    def run_one_step(self):
        self.steps += 1
        for i in range(len(self.array)):
            # most cases
            if i != 0 and i != len(self.array)-1:
                self.array[i] = self.array[i-1] ^ self.array[i+1]
            # corner case: the left-most element
            elif i == 0:
                self.array[i] = self.array[-1] ^ self.array[i+1]
            # corner case: the right-most element
            else:
                self.array[i] = self.array[0] ^ self.array[i-1]
            

    def print_sim_state(self):
        print (f"At step {self.steps}, our state is {self.array}")


rule90 = Rule90(25, [1, 1, 1, 0, 0, 0, 1])
rule90.run()
```

```python
class Counter:
    def __init__(self, low, high):
        self.current = low
        self.high = high

    def __iter__(self):
        return self

    def next(self): 
        if self.current > self.high:
            raise StopIteration
        else:
            self.current += 1
            return self.current - 1

alll = []
for i in range(24): #while True
    for k in Counter(0,11):
        for c in Counter(0,59):
            print ("{}:{}".format(k,c))
            alll.append(("{}:{}".format(k,c)))
```
