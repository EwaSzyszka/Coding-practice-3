```python
'''
__________

Jose
__________

'''

from time import sleep 
# this is not the same time as any of below's



# https://stackoverflow.com/questions/19151/build-a-basic-python-iterator
class ClockIterator():
    def __init__(self, hour = 0, min = 0):
        self.hour = hour
        self.min = min

    def __iter__(self):
        return self

    def __next__(self):

        self.min += 1

        if self.min > 60:
            self.hour += 1
            self.min = 0
        
        if self.hour > 24:
            self.hour = 0

        return str(self.hour) + ":" + str(self.min) 
    

clock = ClockIterator()



for time in clock:
    print(time)
    sleep(0.01)
```
