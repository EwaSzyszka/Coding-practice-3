```python
%%sql

UPDATE housing_listing 
SET     Commission =  CASE  
                 WHEN listing_price < 100000 THEN listing_price * 1.1
                 WHEN listing_price > 100000 AND listing_price < 200000 THEN listing_price * 1.75
                 WHEN listing_price > 200000 AND listing_price < 500000 THEN listing_price * 1.6
                 WHEN listing_price > 500000 AND listing_price < 1000000 THEN listing_price * 1.5
                 WHEN listing_price > 1000000 THEN listing_price * 1.4
                    END      
```

```python
#Updating the sold values based on the buyer table 
```

```python
%%sql 
UPDATE housing_listing
SET sold = 'TRUE'
WHERE EXISTS (SELECT 1 FROM buyer b
                  WHERE b.transaction_id = housing_listing.transaction_id
                 );
```

```python
#calculating the commission
```

```python
%%sql

UPDATE housing_listing 
SET     Commission =  CASE  
                 WHEN listing_price < 100000 THEN listing_price * 1.1
                 WHEN listing_price > 100000 AND listing_price < 200000 THEN listing_price * 1.75
                 WHEN listing_price > 200000 AND listing_price < 500000 THEN listing_price * 1.6
                 WHEN listing_price > 500000 AND listing_price < 1000000 THEN listing_price * 1.5
                 WHEN listing_price > 1000000 THEN listing_price * 1.4
                    END   
```

```python
#Top 5 offices with the most sales for the month 
```

```python
%%sql

CREATE VIEW v_housing_listing as
SELECT *
FROM housing_listing
INNER JOIN buyer ON housing_listing.transaction_id = buyer.transaction_id;
```

```python
#Also, this section is providing details about the buyer
#eg. the sale price/the selling agent 
#more information can be read based on the join on transaction_id's
#the buyer table with the house_listing table based on the transaction id
```

```python
%%sql

SELECT * FROM v_housing_listing
```

```python
%%sql

SELECT mycount,listing_office 
FROM (SELECT listing_office, COUNT(listing_office) mycount 
FROM v_housing_listing
WHERE date_of_sale BETWEEN '2020/02/01' and '2020/02/31' and sold ='TRUE'
GROUP BY listing_office 
ORDER BY COUNT(listing_office) DESC
LIMIT  5
     );
```

```python
#Top 5 estate agents who have sold the most 
```

```python
%%sql
SELECT listing_agent,agent_email,COUNT(listing_agent) sales
FROM v_housing_listing 
WHERE date_of_sale BETWEEN '2020/02/01' and '2020/02/31'  and sold = "TRUE" 
GROUP BY listing_agent
ORDER BY COUNT(listing_agent) DESC
LIMIT    5
```

```python
#Calculate the commission that each estate agent must receive 
```

```python
%%sql
CREATE TABLE Commission_table_
(
  listing_agent TEXT, 

  agreggate_profit INT
); 
```

```python
%%sql
INSERT INTO Commission_table_ (listing_agent,agreggate_profit)
SELECT listing_agent,sum(Commission -listing_price) as agreggate_profit 
FROM v_housing_listing 
WHERE date_of_sale 
BETWEEN '2020/02/01' and '2020/02/31'   and sold = "TRUE" 
GROUP BY listing_agent
ORDER BY agreggate_profit DESC
```

```python
%%sql

SELECT * FROM Commission_table_
```

```python
#For all houses that were sold that month, calculate the average number of days 
#that the house was on the market.
```

```python
%%sql 
SELECT transaction_id, date_of_listing,date_of_sale
FROM v_housing_listing 
WHERE date_of_sale 
BETWEEN '2020/02/01' and '2020/02/31'
```

```python
%%sql
SELECT julianday(replace(date_of_sale, '/', '-')) - julianday(replace(date_of_listing, '/', '-')) as difference_in_number_of_days
FROM v_housing_listing
```

```python
#For all houses that were sold that month, calculate the average selling price
```

```python
%%sql

SELECT sum(Commission)/COUNT(Commission) as average_price FROM v_housing_listing WHERE date_of_sale 
BETWEEN '2020/02/01' and '2020/02/31' and sold='TRUE'
```

```python
#Find the zip codes with the top 5 average sales prices
#Commission is a column in the table that has the price with the commission included calculated
```

```python
%%sql

SELECT transaction_id,zip_code,Commission as average_price FROM v_housing_listing WHERE date_of_sale 
BETWEEN '2020/02/01' and '2020/02/31' and sold='TRUE'
ORDER BY average_price DESC
LIMIT    5
```
